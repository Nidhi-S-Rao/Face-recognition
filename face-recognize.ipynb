```python
import numpy as np
import cv2
import pickle
labels={"person_name":1}
face_cascade=cv2.CascadeClassifier('cascades/data/haarcascade_frontalface_alt2.xml')
#eye_cascade=cv2.CascadeClassifier('C:\\Users\\HP\\Desktop\\Projects\\ML\\cascades\\data\\haarcascade_eye.XML')


recognizer = cv2.face.LBPHFaceRecognizer_create()
recognizer.read("Trainer.yml")
with open("labels.pickle", 'rb') as f:
    og_labels=pickle.load(f)
    labels={v:k for k,v in og_labels.items()}
```

```python
#(camx,camy)=(320,240)
cam=cv2.VideoCapture(0)
#cam.set(3,camx)
#cam.set(4,camy)
```

```python
while(True):
    ret,frame=cam.read()
    gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    faces=face_cascade.detectMultiScale(gray,scaleFactor=1.5, minNeighbors=5)
    
    for(x,y,w,h) in faces:
        #print(x,y,w,h)
        #cv2.rectangle(frame, (x, y), (x1, y1), (255, 0, 0), 2)
        roi_gray=gray[y:y+h, x:x+w]
        roi_color=frame[y:y+h, x:x+w]
        id_,conf=recognizer.predict(roi_gray)
        if conf>=45:
            k=labels[id_]
            print(id_)
            print(labels[id_])
            font=cv2.FONT_HERSHEY_SIMPLEX
            name=labels[id_]
            color=(255,0,0) #BGR
            stroke=1
            cv2.rectangle(frame,(x,y),(x+w,y+h),color,stroke)
            color=(255,255,255)
            stroke=3
            cv2.putText(frame,name,(x,y),font,1,color,stroke,cv2.LINE_AA)
            #eyes=eye_cascade.detectMultiScale(roi_gray)
            #for (ex,ey,ew,eh) in eyes:
                #cv2.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(0,255,0),stroke)
        
        
        
        
           
       
   







    cv2.imshow('frame',frame)
    if(cv2.waitKey(1)==ord('q')):
        img_itemG=k+"G.png"
        img_itemC= k+"C.png"
        cv2.imwrite(img_itemG,roi_gray)
        cv2.imwrite(img_itemC,roi_color)
        break
        
cam.release()
cv2.destroyAllWindows()

        
```
